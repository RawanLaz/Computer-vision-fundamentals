```python
import cv2
from google.colab.patches import cv2_imshow # for image display
from skimage import io
import numpy as np
from matplotlib import pyplot as plt
```

# Downloading and showing images




```python

# Create a list to store the urls of the images
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg",
       "https://iiif.lib.ncsu.edu/iiif/0016007/full/800,/0/default.jpg",
        "https://raw.githubusercontent.com/opencv/opencv/3.4/samples/data/LinuxLogo.jpg",
        "https://raw.githubusercontent.com/opencv/opencv/3.4/samples/data/WindowsLogo.jpg"]  

        
images = []

# Read and display the image
# loop over the image URLs, you could store several image urls in the list

for url in urls:
  image = io.imread(url)
  image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
  images.append(image)
      
  cv2_imshow(image)
```

# Convolve image with a kernal with an image (Image Filtering)
https://docs.opencv.org/master/d4/d13/tutorial_py_filtering.html

Screenshot from 2021-06-02 10-17-42.png

```python
image = images[3].copy()


kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(image,-1,kernel)
print ("filter 2D")
combined_images = np.hstack((image, dst))
cv2_imshow(combined_images)

# or u can use the opencv function cv.blur(img,(5,5))
dst = cv2.blur(image,(5,5))
print ("blur")
combined_images = np.hstack((image, dst))
cv2_imshow(combined_images)
```

Note that they produced similar results

# Gaussian filter 

```python
image = images[3].copy()

gblur = cv2.GaussianBlur(image,(5,5),0) # is highly effective in removing Gaussian noise from an image.
combined_images = np.hstack((image, gblur))
cv2_imshow(combined_images)

median = cv2.medianBlur(image,5) # This is highly effective against salt-and-pepper noise in an image
combined_images = np.hstack((image, median))
cv2_imshow(combined_images)

bilblur = cv2.bilateralFilter(image,9,75,75) # is highly effective in noise removal while keeping edges sharp
combined_images = np.hstack((image, bilblur))
cv2_imshow(combined_images)
```

```python
import cv2
from google.colab.patches import cv2_imshow # for image display
from skimage import io
import numpy as np
from matplotlib import pyplot as plt
```

# Downloading and showing images


```python
# Create a list to store the urls of the images
urls = ["https://iiif.lib.ncsu.edu/iiif/0052574/full/800,/0/default.jpg"]  

images = []

# Read and display the image
# loop over the image URLs, you could store several image urls in the list

for url in urls:
  image = io.imread(url)
  gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

  images.append(gray)
      
  cv2_imshow(gray)
```

# Generate Gaussian noise

http://datahacker.rs/opencv-common-types-of-noise/

```python
image = images[0].copy()

gaussian_noise = np.zeros((image.shape[0], image.shape[1]),dtype=np.uint8)
# set a mean value to 128 and a standard deviation to 20. Note that the larger 
# std -> the more noise values
cv2.randn(gaussian_noise, 128, 20)

noisy_image1 = cv2.add(image,gaussian_noise)


cv2_imshow(gaussian_noise)
cv2_imshow(noisy_image1)
```
